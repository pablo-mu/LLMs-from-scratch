## 2.2 Tokenizing Text

```python
import urllib.request
url = ("https://raw.githubusercontent.com/rasbt/"
"LLMs-from-scratch/main/ch02/01_main-chapter-code/"
"the-verdict.txt")
file_path = "the-verdict.txt"
urllib.request.urlretrieve(url, file_path)
```

Reading a short story as text sample

```python
with open("the-verdict.txt", "r", encoding = "utf-8") as f:
    raw_text = f.read()
print("Total number of character:", len(raw_text))
print(raw_text[:99])
```

Now, we are going to tokenize, which is the process of breaking a text into words, phrases, symbols, or other meaningful elements. In this case, in this case, we build a tokenizer using the regular expression module `re` to split the text into words.

```python
import re
preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)
# Now we remove whitespaces.
preprocessed = [item.strip() for item in preprocessed if item.strip()]
print(len(preprocessed))
```
